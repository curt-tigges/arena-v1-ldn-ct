```python
import torch as t
```

```python
import sys
sys.path.append('../common_modules')
```

```python
from sample_methods import sample_tokens
from nlp_modules import WordsTokenizer
```

```python
def generate(input):

    MODEL_FILENAME = "./w1d3_transformer_shakespeare.pt"

    model = t.load(MODEL_FILENAME, map_location=t.device('cpu'))
    model.eval()

    tokenizer = WordsTokenizer(16)
    tokenizer.load_saved()

    text_output = sample_tokens(model, tokenizer, input, max_tokens_generated=100, temperature=1.0, top_k=10)
    return text_output
```

```python
import gradio as gr

def greet(name):
    return "Hello " + name + "!"

demo = gr.Interface(fn=generate, inputs="text", outputs="text")

demo.launch(debug=True)
```
